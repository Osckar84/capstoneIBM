Above coordinates are just plain numbers that can not give you any intuitive insights about where are those launch sites. If you are very good at geography, you can interpret those numbers directly in your mind. If not, that's fine too. Let's visualize those locations by pinning them on a map.


We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.


```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
```

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,


```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
# Función para calcular distancia con Haversine
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Radio de la Tierra en km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    return R * c

# Definir el centro del mapa con la coordenada promedio de los sitios de lanzamiento
nasa_coordinate = [launch_sites_df["Lat"].mean(), launch_sites_df["Long"].mean()]

# Crear el mapa base
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Crear un grupo de clusters para los marcadores
marker_cluster = MarkerCluster().add_to(site_map)

# Coordenadas de la costa más cercana (ajústalas con los valores reales)
coastline_lat = 28.56367  
coastline_lon = -80.57163  

# Coordenadas de puntos cercanos (ajusta con valores reales obtenidos con MousePosition)
closest_city = [28.3922, -80.6077]  # Ejemplo: ciudad cercana
closest_highway = [28.563, -80.576]  # Ejemplo: carretera cercana
closest_railway = [28.398, -80.605]  # Ejemplo: vía de tren

# Diccionario con las ubicaciones cercanas
proximities = {
    "Ciudad": closest_city,
    "Autopista": closest_highway,
    "Ferrocarril": closest_railway,
    "Costa": [coastline_lat, coastline_lon]
}

# Agregar marcadores y líneas de conexión al mapa
launch_site_lat = launch_sites_df.iloc[0]["Lat"]
launch_site_lon = launch_sites_df.iloc[0]["Long"]
launch_site_coord = [launch_site_lat, launch_site_lon]

for name, coord in proximities.items():
    distance = calculate_distance(launch_site_lat, launch_site_lon, coord[0], coord[1])

    # Crear marcador con distancia
    folium.Marker(
        location=coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{name}: {distance:.2f} KM</b></div>'
        )
    ).add_to(site_map)

    # Dibujar línea entre el sitio de lanzamiento y el punto cercano
    folium.PolyLine(
        locations=[launch_site_coord, coord],
        weight=2,
        color="blue",
        opacity=0.8
    ).add_to(site_map)

# Agregar marcadores de lanzamientos al mapa
for _, launch in spacex_df.iterrows():
    marker_color = "green" if launch["class"] == 1 else "red"
    
    folium.Marker(
        location=[launch["Lat"], launch["Long"]],
        popup=f"{launch['Launch Site']} - {'Success' if launch['class'] == 1 else 'Failed'}",
        icon=folium.Icon(color=marker_color)
    ).add_to(marker_cluster)

# Agregar la línea del ecuador
folium.PolyLine(
    [[0, -180], [0, 180]], color="blue", weight=2, opacity=0.8
).add_to(site_map)

# Agregar MousePosition para ver coordenadas en tiempo real
MousePosition(
    position="topright",
    separator=" | ",
    prefix="Coordenadas:",
    lat_formatter=lambda x: f"{x:.5f}",
    lng_formatter=lambda x: f"{x:.5f}"
).add_to(site_map)

# Mostrar el mapa con todas las mejoras
site_map
  
```

## 1. ¿Todos los sitios de lanzamiento están cerca del ecuador?
No necesariamente. Aunque muchos sitios de lanzamiento están ubicados relativamente cerca del ecuador, no todos lo están.
- Beneficio del ecuador: Los sitios cercanos al ecuador aprovechan la rotación de la Tierra, lo que les proporciona un impulso adicional y reduce el combustible necesario para alcanzar la órbita.
- Ejemplo lejano: Algunos sitios, como Vandenberg Air Force Base (California, EE.UU.), están mucho más alejados del ecuador porque se especializan en órbitas polares y heliosíncronas, ideales para observación terrestre y misiones científicas.
## 2. ¿Todos los sitios de lanzamiento están cerca de la costa?
Sí, la mayoría de los sitios de lanzamiento están cerca de costas o grandes cuerpos de agua.
- Seguridad: Esta ubicación reduce el riesgo de caída de escombros o fallos de lanzamiento sobre áreas pobladas.
- Trayectorias favorables: Lanzar sobre el océano permite acceder a diferentes tipos de órbitas, especialmente las que requieren un desplazamiento hacia el este, aprovechando la rotación terrestre.
## 3. Explicación de los hallazgos
- Los sitios cerca del ecuador son ideales para órbitas geoestacionarias, donde los satélites se mantienen sobre un punto fijo de la Tierra.
- Los sitios costeros favorecen la seguridad y permiten trayectorias de lanzamiento sin riesgo para poblaciones cercanas.
- Los sitios lejos del ecuador como Vandenberg están optimizados para órbitas polares, esenciales para misiones de observación terrestre y satélites meteorológicos.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

 # Análisis de proximidades de los sitios de lanzamiento
## 1️. ¿Los sitios de lanzamiento están cerca de vías de tren?
No es una regla general, aunque encontramos algunos puntos cerca:
- Las conexiones ferroviarias pueden ser útiles para transportar partes de cohetes grandes.
- Algunos sitios tienen líneas de tren dedicadas, mientras que otros dependen más de carreteras para logística.
## 2️. ¿Los sitios de lanzamiento están cerca de autopistas?
Sí, la mayoría de los sitios están cerca de carreteras principales.
- Las autopistas facilitan el transporte de cohetes, combustible y equipos.
- La presencia de amplias carreteras asegura el traslado seguro de cargas sensibles.
## 3️. ¿Los sitios de lanzamiento están cerca de la costa?
Sí, casi todos los sitios de lanzamiento están cerca del mar.
- Esto es intencional, ya que lanzar sobre el océano reduce riesgos por caída de escombros en fallos.
- Además, permite zonas de caída seguras para las etapas de los cohetes que se separan durante el vuelo.
## 4️. ¿Los sitios de lanzamiento mantienen distancia de las ciudades?
Sí, los sitios de lanzamiento están estratégicamente ubicados lejos de zonas urbanas.
- Esto minimiza riesgos en caso de explosiones o fallos de lanzamiento.
- Algunos están en desiertos o islas remotas, garantizando máxima seguridad.

# Explicación de los hallazgos
1. Ferrocarriles y autopistas: fundamentales para la logística, aunque las carreteras tienen mayor prioridad.
2. Ubicación costera: medida de seguridad para minimizar riesgos en lanzamientos.
3. Distancia de ciudades: precaución esencial para evitar problemas en áreas pobladas.
